```python
python_df=pd.concat([country_uses_pytho,countr_respondent],axis=1,sort=False)
python_df
```

```python
python_df=python_df.rename(columns={'Country':'Num_of Resp','LanguageHaveWorkedWith':'Language_by_people' })
```

```python
python_df['%_of_know_python']=(python_df['Language_by_people']/python_df['Num_of Resp'])*100
```
